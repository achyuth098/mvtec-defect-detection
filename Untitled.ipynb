```python
import os
from glob import glob
import random
import matplotlib.pyplot as plt
from PIL import Image
```

```python
bottle_path = "Data/bottle/"
leather_path = "Data/leather/"
```

```python
def count_images(category_path):
    train_good = glob(os.path.join(category_path, "train", "good", "*.png"))
    test_good = glob(os.path.join(category_path, "test", "good", "*.png"))
    defect_types = [d for d in os.listdir(os.path.join(category_path, "test")) if d != "good"]
    test_defect_counts = {}
    for defect in defect_types:
        images = glob(os.path.join(category_path, "test", defect, "*.png"))
        test_defect_counts[defect] = len(images)
    print(f"Train Good: {len(train_good)}")
    print(f"Test Good: {len(test_good)}")
    for defect, count in test_defect_counts.items():
        print(f"Test {defect}: {count}")
```

```python
def show_sample_images(category_path, n=3):
    good_imgs = glob(os.path.join(category_path, "train", "good", "*.png"))
    plt.figure(figsize=(12, 4))
    for i, img_path in enumerate(random.sample(good_imgs, n)):
        img = Image.open(img_path)
        plt.subplot(2, n, i+1)
        plt.imshow(img)
        plt.axis('off')
        plt.title("Normal")
    defect_types = [d for d in os.listdir(os.path.join(category_path, "test")) if d != "good"]
    defect_imgs = glob(os.path.join(category_path, "test", defect_types[0], "*.png"))
    for i, img_path in enumerate(random.sample(defect_imgs, n)):
        img = Image.open(img_path)
        plt.subplot(2, n, n+i+1)
        plt.imshow(img)
        plt.axis('off')
        plt.title(defect_types[0])
    plt.show()
```

```python
print("Bottle:")
count_images(bottle_path)
show_sample_images(bottle_path)
```

```python
print("Leather:")
count_images(leather_path)
show_sample_images(leather_path)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
```

```python
# Set device (use GPU if available)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# Set image size
IMG_SIZE = 128

# Define transformations
transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
])

# Custom dataset class for loading images
from PIL import Image
import glob

class MVTecDataset(Dataset):
    def __init__(self, image_dir, transform=None):
        self.image_paths = glob.glob(os.path.join(image_dir, "*.png"))
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img = Image.open(self.image_paths[idx]).convert("RGB")
        if self.transform:
            img = self.transform(img)
        return img
```

```python
# Change these paths as needed
bottle_good_train_dir = "Data/bottle/train/good/"
leather_good_train_dir = "Data/leather/train/good/"

# Load only normal (good) images for training
bottle_train_dataset = MVTecDataset(bottle_good_train_dir, transform=transform)
bottle_train_loader = DataLoader(bottle_train_dataset, batch_size=32, shuffle=True)

leather_train_dataset = MVTecDataset(leather_good_train_dir, transform=transform)
leather_train_loader = DataLoader(leather_train_dataset, batch_size=32, shuffle=True)

print("Bottle train samples:", len(bottle_train_dataset))
print("Leather train samples:", len(leather_train_dataset))
```

```python
class SimpleAutoencoder(nn.Module):
    def __init__(self):
        super(SimpleAutoencoder, self).__init__()
        # Encoder
        self.encoder = nn.Sequential(
            nn.Conv2d(3, 16, 3, stride=2, padding=1),  # [B, 16, 64, 64]
            nn.ReLU(),
            nn.Conv2d(16, 32, 3, stride=2, padding=1), # [B, 32, 32, 32]
            nn.ReLU(),
            nn.Conv2d(32, 64, 3, stride=2, padding=1), # [B, 64, 16, 16]
            nn.ReLU(),
        )
        # Decoder
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(64, 32, 3, stride=2, padding=1, output_padding=1), # [B, 32, 32, 32]
            nn.ReLU(),
            nn.ConvTranspose2d(32, 16, 3, stride=2, padding=1, output_padding=1), # [B, 16, 64, 64]
            nn.ReLU(),
            nn.ConvTranspose2d(16, 3, 3, stride=2, padding=1, output_padding=1),  # [B, 3, 128, 128]
            nn.Sigmoid()
        )

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x

autoencoder = SimpleAutoencoder().to(device)
```

```python
# Training loop for bottle data
num_epochs = 10
criterion = nn.MSELoss()
optimizer = optim.Adam(autoencoder.parameters(), lr=0.001)

autoencoder.train()
for epoch in range(num_epochs):
    epoch_loss = 0
    for imgs in bottle_train_loader:
        imgs = imgs.to(device)
        outputs = autoencoder(imgs)
        loss = criterion(outputs, imgs)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss/len(bottle_train_loader):.4f}")
```

```python
# Directory containing only "good" training images for leather
leather_good_train_dir = "Data/leather/train/good/"

# Create dataset and dataloader for leather
leather_train_dataset = MVTecDataset(leather_good_train_dir, transform=transform)
leather_train_loader = DataLoader(leather_train_dataset, batch_size=32, shuffle=True)

print("Number of good leather images:", len(leather_train_dataset))
```

```python
# Initialize a new autoencoder for leather
leather_autoencoder = SimpleAutoencoder().to(device)

num_epochs = 10
criterion = nn.MSELoss()
optimizer = optim.Adam(leather_autoencoder.parameters(), lr=0.001)

leather_autoencoder.train()
for epoch in range(num_epochs):
    epoch_loss = 0
    for imgs in leather_train_loader:
        imgs = imgs.to(device)
        outputs = leather_autoencoder(imgs)
        loss = criterion(outputs, imgs)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss/len(leather_train_loader):.4f}")
```

```python
torch.save(autoencoder.state_dict(), "bottle_autoencoder.pth")
torch.save(leather_autoencoder.state_dict(), "leather_autoencoder.pth")
```

```python
# Test directories for bottle
bottle_good_test_dir = "Data/bottle/test/good/"
bottle_defect_dirs = [os.path.join("Data/bottle/test", d) 
                      for d in os.listdir("Data/bottle/test") if d != "good"]

# Load normal test images
bottle_good_test_dataset = MVTecDataset(bottle_good_test_dir, transform=transform)
bottle_good_test_loader = DataLoader(bottle_good_test_dataset, batch_size=1, shuffle=False)

# Load all defect images
bottle_defect_image_paths = []
for d in bottle_defect_dirs:
    bottle_defect_image_paths.extend(glob.glob(os.path.join(d, "*.png")))
print("Number of good test images:", len(bottle_good_test_dataset))
print("Number of defect test images:", len(bottle_defect_image_paths))
```

```python
#For Leather (Repeat the Above with Paths Changed)

# Test directories for leather
leather_good_test_dir = "Data/leather/test/good/"
leather_defect_dirs = [os.path.join("Data/leather/test", d) 
                       for d in os.listdir("Data/leather/test") if d != "good"]

# Load normal test images
leather_good_test_dataset = MVTecDataset(leather_good_test_dir, transform=transform)
leather_good_test_loader = DataLoader(leather_good_test_dataset, batch_size=1, shuffle=False)

# Load all defect images
leather_defect_image_paths = []
for d in leather_defect_dirs:
    leather_defect_image_paths.extend(glob.glob(os.path.join(d, "*.png")))
print("Number of good test images:", len(leather_good_test_dataset))
print("Number of defect test images:", len(leather_defect_image_paths))
```

```python
#Helper Function: Compute Error for One Image

import numpy as np

def get_reconstruction_error(model, img_tensor):
    model.eval()
    with torch.no_grad():
        img_tensor = img_tensor.unsqueeze(0).to(device) # add batch dimension
        output = model(img_tensor)
        error = torch.mean((output - img_tensor) ** 2).item()
    return error, output.squeeze(0).cpu()
```

```python
#Calculate Errors for Normal and Defect Images
# Example: bottle
normal_errors = []
for img in bottle_good_test_dataset:
    err, _ = get_reconstruction_error(autoencoder, img)
    normal_errors.append(err)
print("Mean reconstruction error (normal):", np.mean(normal_errors))
```

```python
#Defect Images

def load_img_as_tensor(img_path, transform):
    img = Image.open(img_path).convert("RGB")
    return transform(img)

defect_errors = []
for img_path in bottle_defect_image_paths:
    img_tensor = load_img_as_tensor(img_path, transform)
    err, _ = get_reconstruction_error(autoencoder, img_tensor)
    defect_errors.append(err)
print("Mean reconstruction error (defect):", np.mean(defect_errors))
```

```python
#Visualize the Difference: Histogram
plt.hist(normal_errors, bins=30, alpha=0.7, label='Normal')
plt.hist(defect_errors, bins=30, alpha=0.7, label='Defect')
plt.xlabel('Reconstruction Error')
plt.ylabel('Count')
plt.legend()
plt.title('Reconstruction Error: Normal vs Defect (Bottle)')
plt.show()
```

```python
def show_reconstruction(model, img_path, transform):
    img = Image.open(img_path).convert("RGB")
    img_tensor = transform(img)
    model.eval()
    with torch.no_grad():
        recon = model(img_tensor.unsqueeze(0).to(device))
        recon = recon.squeeze(0).cpu()
    # Plot
    fig, axes = plt.subplots(1, 3, figsize=(12, 4))
    axes[0].imshow(np.transpose(img_tensor.numpy(), (1,2,0)))
    axes[0].set_title("Original")
    axes[0].axis('off')
    axes[1].imshow(np.transpose(recon.numpy(), (1,2,0)))
    axes[1].set_title("Reconstruction")
    axes[1].axis('off')
    diff = (np.abs(img_tensor.numpy() - recon.numpy())).mean(axis=0)
    im = axes[2].imshow(diff, cmap='hot')
    axes[2].set_title("Difference (Error map)")
    axes[2].axis('off')
    fig.colorbar(im, ax=axes[2])
    plt.show()

# Example for a normal image
show_reconstruction(autoencoder, bottle_good_test_dataset.image_paths[0], transform)
# Example for a defect image
show_reconstruction(autoencoder, bottle_defect_image_paths[0], transform)
```

```python
# Example: Calculate threshold from normal errors (mean + 2*std is common)
threshold = np.mean(normal_errors) + 2 * np.std(normal_errors)
print("Defect detection threshold:", threshold)

# Test how many defects are caught
detected_defects = [e > threshold for e in defect_errors]
print("Defect detection rate: {:.2f}%".format(100 * np.mean(detected_defects)))
```
